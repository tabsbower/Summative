6. Filter the DataFrame on **one specific module** (your choice) and plot its **attendance rate over time**

### **Exercise 5 â€” Attendance Relationships**



1. Answer the question: **Do students with low overall attendance tend to attend modules that have low attendance?**.

2. **[For L5 students only]**. Choose between a correlation analysis or a linear regression analysis to answer the question.

3. **[For L6 students only]** Run both correlation analysis and linear regression analysis to answer the same question. Do they tell the same story?

> Tips:
> 1. Compute the mean attendance rate of each module. Can you re-use some of the `groupby` code you wrote above?
> 2. For each module, compute the average overall attendance of the students who attended that module.
> 3. Correlation and linear regression can both help you answer the question. 
> 4. Report main statistic and p-value(s).


````python
#compute mean attendance rate per module
module_means = attendance_df_new.groupby('Module Name')['Attended'].mean().reset_index().rename(columns={"Attended": "ModuleMeanAttendance"})

#compute average overall attendance per student
module_overall = attendance_df_new.groupby('Module Name')['Student Overall Attendance'].mean().reset_index().rename(columns={"Student Overall Attendance": "Average Student Overall"})

#merge into one dataframe
analysis_df = module_means.merge(module_overall, on="Module Name")

#correlation analysis
r, p = pearsonr(analysis_df["ModuleMeanAttendance"], analysis_df["Average Student Overall"])

print("Correlation:", r)
print("p-value:", p)

#linear regression analysis
import statsmodels.api as sm

X = sm.add_constant(analysis_df["Average Student Overall"])
y = analysis_df["ModuleMeanAttendance"]

model = sm.OLS(y, X).fit()
print(model.summary())
````

For the correlation analysis, a correlation coefficient closer to 1 suggests a positive relationship and a low p value suggests that the hypothesis is significant. In the case of this question it suggests a strong positive relationship between students with low overall attendance and their tendency to take modules with low overall attendance. From the linear regression we also see a very low p value (p<0.001) and a slope with gradient close to 1. This suggests the same as the correlation analysis. The two analyses support the conclusion that students with low overall attendance tend to attend modules that have lower attendance.
